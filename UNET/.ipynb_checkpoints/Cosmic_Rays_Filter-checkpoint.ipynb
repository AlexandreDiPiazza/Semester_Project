```python
from scipy.ndimage.filters import laplace
def Remove_Cosmic_Rays(im) : 
    im_filtered = im.copy()
    count = 0
    Laplace = laplace(im)
    extreme = np.mean(Laplace) - np.std(Laplace)
    for i in range(np.shape(im)[0]) : 
        for j in range(np.shape(im)[1]) : 
            if im[i,j] == 1 and Laplace[i,j] < extreme: 
                count += 1
                im_filtered[i,j] = np.mean(im_filtered[i-10:i+10, j-10:j+10]) 
    print('Number of pixels changed : ' + str(count))
    return im_filtered
```

```python
im1_filtered = Remove_Cosmic_Rays(im1)
```
